```python
!pip install pydub

import os
from pydub import AudioSegment

def convert_all_wav_to_mp3(input_folder=".", output_folder="output", bitrate="128k"):
    # 出力フォルダを作成（存在しない場合）
    os.makedirs(output_folder, exist_ok=True)

    # フォルダ内のすべてのWAVファイルを処理
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(".wav"):
            input_wav = os.path.join(input_folder, filename)
            output_mp3 = os.path.join(output_folder, os.path.splitext(filename)[0] + ".mp3")

            # WAVを読み込み
            audio = AudioSegment.from_wav(input_wav)

            # MP3として保存
            audio.export(output_mp3, format="mp3", bitrate=bitrate, parameters=["-q:a", "7", "-ar", "44100"])

            print(f"変換完了: {output_mp3}")

# 実行（カレントフォルダのWAVをすべて変換）
convert_all_wav_to_mp3()

import shutil # import the shutil module

# Define OUTPUT_FOLDER and ZIP_FILENAME
OUTPUT_FOLDER = "output"  # or any other desired folder name
ZIP_FILENAME = "DownloadThis"  # or any other desired filename

# Create the archive
shutil.make_archive(ZIP_FILENAME, 'zip', OUTPUT_FOLDER)

# Display the path of the ZIP file
ZIP_PATH = ZIP_FILENAME + ".zip"  # Add the .zip extension
print(f"ZIP file created at: {ZIP_PATH}")
```
